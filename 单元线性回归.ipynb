```python
#from sklearn.datasets import load_boston
import numpy as np
import matplotlib.pyplot as plt
from superai.regression.model.linear_regression import Linear_Regression, polynomial_expand
```

```python
# 简单一元线性回归
X = np.linspace(0, 1, 20)
salt = np.random.rand(20) - 0.5
Y = 10*X + 6 + salt
plt.scatter(X, Y)
plt.figure()
```

```python
model = Linear_Regression(1)
model.fit(X[np.newaxis, :], Y[np.newaxis, :], 0.1, verbose=False)
# print(model.w)
# print(model.b)

X_test = np.linspace(0, 1, 100)
Y_test = model.predict(X_test[np.newaxis, :])

plt.scatter(X, Y)
plt.plot(X_test, Y_test.flatten(), color="red")
plt.figure()
```

```python
# 一元多项式线性回归

X = np.linspace(0, 2.5, 100)
Y = 3 * X**2 - 2 * X + 3
plt.plot(X, Y, '.')
plt.figure()
```

```python
model = Linear_Regression(1)
model.fit(X[np.newaxis, :], Y[np.newaxis, :], 0.1, 2000, verbose=False)
Y_pre = model.predict(X)
plt.plot(X, Y, '.')
plt.plot(X, Y_pre.flatten())
plt.figure()
```

```python
X_poly = polynomial_expand(X, 2)
min_ = np.min(X_poly, axis=1, keepdims=True)
max_ = np.max(X_poly, axis=1, keepdims=True)
X_poly = (X_poly - min_) / ((max_ - min_) + 1e-32)

model = Linear_Regression(2)
model.fit(X_poly, Y, 0.1, 10000, verbose=False)

Y_pre = model.predict(X_poly)
plt.plot(X, Y_pre.flatten(), '.')
plt.plot(X, Y, '-')
plt.figure()
```
