```python
from sklearn.datasets import load_boston
import numpy as np
import matplotlib.pyplot as plt
from superai.regression.model.linear_regression import Linear_Regression
```

```python
# 多元线性回归--波士顿房价预测问题
boston = load_boston()

train_data = boston['data']


train_data = train_data.transpose()

# 归一化
min_ = np.min(train_data, axis=1, keepdims=True)
max_ = np.max(train_data, axis=1, keepdims=True)
train_data = (train_data - min_) / ((max_ - min_) + 1e-32)


# 标签
label = boston['target']
label = label[np.newaxis, :]

model = Linear_Regression(13)
model.fit(train_data, label, 0.1, iteration_count=1000, verbose = False)

label_pre = model.predict(train_data)

X = np.linspace(0, 1000, 506)

plt.subplot(121)
plt.xlim(-10, 1100)
plt.ylim(-10, 60)
plt.scatter(X, label_pre.flatten(),  color="red")
plt.subplot(122)
plt.xlim(-10, 1100)
plt.ylim(-10, 60)
plt.scatter(X, label.flatten(),  color="blue")
plt.figure()

```
